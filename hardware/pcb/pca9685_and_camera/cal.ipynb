# TPS61288 calculations

## Inductor

```python
Vins = [3, 5.2]
Vout = 6
Iout = 1
L = 1e-6
fsw = 500e3

servo_current_max = 300e-3

max_servos_same= 4

Iout_max = max_servos_same*servo_current_max
stall_current = 0.8+3*servo_current_max


Ceff = 0.8
results ={}
for Vin in Vins:
    Idc = Vout*Iout_max/(Vin*Ceff)
    results[Vin]={"Idc": Idc, "Idc_stall":Vout*stall_current/(Vin*Ceff)}

    print(f"{Vin} {results[Vin]}")

def ipp_calc(L, Vout, Vin, fsw):
    return(1/(L*((1/(Vout-Vin)+1/Vin)*fsw)))

for Vin, v in results.items():
    v["Ipp"] = ipp_calc(L, Vout, Vin, fsw)
    v["Ilpeak"] = v["Idc"]+v["Ipp"]/2
    v["Ilpeak_stall"] = v["Idc_stall"]+v["Ipp"]/2
    print (f"{Vin}: {repr(v)}")
```
